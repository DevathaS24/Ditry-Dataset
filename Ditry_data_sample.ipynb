```python
# Importing pandas library
import pandas as pd
```

```python
# Reading the data set
dirty = pd.read_excel("/content/Ditry Data Sample.xlsx")
dirty
```

# Consumer Part Data Cleaning

# Consumer First Class

```python
# Customer first class
cfc = dirty.loc[range(1,824),['Segment>>','Consumer']]
cfc
```

```python
# Drop the second column which has OrderID
cfc = cfc.drop(cfc.index[0])
cfc
```

```python
# Drop all the NaN values in Consumer column
cfc = cfc.dropna()
cfc
```

```python
# add the Segment column as the first column
cfc.insert(0,"Segment","Consumer")
cfc
```

```python
# add the ship mode
cfc.insert(1,"Ship Mode","First Class")
```

```python
cfc
```

```python
cfc.columns
```

```python
# rename the columns name
cfc = cfc.rename({'Segment>>':"Order ID",'Consumer':"Sales"},axis = 1)
cfc
```

```python
# to calculate the sum of sales total for the Customer sector
cfc.loc[len(cfc["Segment"])] = ["Consumer" , "First Class" , "Grand Total" , cfc["Sales"].sum()]
cfc
```

```python
cfc = cfc.reset_index()
cfc
```

```python
cfc = cfc.drop(["index"],axis = 1)
cfc
```
